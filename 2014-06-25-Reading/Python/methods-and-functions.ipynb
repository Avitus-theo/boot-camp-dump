**Exercise**

Take the code below and make it into a function that takes one argument - the name of the file to be analyzed.  

```python
import numpy as np
from matplotlib import pyplot as plt

def analyse(file_name):
    """Takes a file as input and analyses it."""
    data = np.loadtxt(fname=file_name, delimiter=',')

    plt.figure(figsize=(10.0, 3.0))

    plt.subplot(1, 3, 1)
    plt.ylabel('average')
    plt.plot(data.mean(0))

    plt.subplot(1, 3, 2)
    plt.ylabel('max')
    plt.plot(data.max(0))

    plt.subplot(1, 3, 3)
    plt.ylabel('min')
    plt.plot(data.min(0))

    plt.tight_layout()
    plt.show()

analyse('inflammation-01.csv')
```
